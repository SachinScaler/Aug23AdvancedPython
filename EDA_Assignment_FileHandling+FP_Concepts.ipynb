````python
path = '/content/sample_data/california_housing_test.csv'
with open(path, 'r') as file:
  all_data = file.readlines()

print("Total Rows", len(all_data))
print(all_data)
````

````python
# first row in data is column names/features and second row onwards is values all in string format.
columns=all_data[0]
data = all_data[1:]

# let's split on comma and print all columns and first row of data
print(len(columns.split(",")))
print(len(data[0].split(",")))


print(columns.split(","))
print(data[0].split(","))
````

- as expected after splitting both column and row have 9 values indicating one to one relationship between col and rows

````python
### Let's utilize a dictionary to store and access data columnwise. You can use any other Data Structure/simple list as well
dict_data= {col_name: [] for col_name in columns.split(",")}
dict_data
````

````python
# Let's fill the values in this dictionary

# For understanding what's happening here
for row in data:
  row = list(map(float, row.split(","))) # why mapping remaining data to float?
  for index,key in enumerate(dict_data):
    print(index, key, row[index])
  break

# everythinng looks good and we are able to access column and values. Let's use this to fill the dictionary
````

````python
# Let's fill the values in this dictionary
for row in data:
  row = list(map(float, row.split(","))) # why mapping remaining data to float?
  for index,key in enumerate(dict_data):
    dict_data[key].append(row[index])
print(dict_data)
````
